1. Write a program to check if number is Armstrong.


```python
number = input("Enter Number to check if Number is Armstrong: ")
len_num = len(number)
sum = 0
for i in range(len_num):
  sum += int(number[i])**len_num

if sum == int(number):
  print("Number is Armstrong")
else:
  print("Number is not Armstrong")
```

2. Write a program to check special number. (Number is equal to the sum of its divisors).

```python
def fact(n):
  factorial = 1
  for i in range(n):
    factorial *= n - i
  return factorial

number = input("Enter Number to check if Number is Special: ")
len_num = len(number)
sum = 0
for i in range(len_num):
  sum += fact(int(number[i]))

if sum == int(number):
  print("Number is Special")
else:
  print("Number is not Special")
```

```python
number = int(input("Enter Number to check if Number is Perfect: "))
sum = 0
for i in range(1,number):
  if number % i == 0:
    sum += i

if sum == int(number):
  print("Number is Perfect")
else:
  print("Number is not Perfect")
```

7. Write a program to count number of digits, upper-case characters and lower-case characters from the string.

```python
st= input("Enter String :")
# 'I love doing python 134 @# programming in spyder'
count_num = 0
count_lower = 0
count_upper = 0
count_symbol = 0
for i in range(len(st)):
  if st[i] == ' ':
    continue
  elif st[i] >= '0' and st[i] <= '9':
    count_num += 1
  elif st[i] >= 'A' and st[i] <= 'Z':
    count_upper += 1
  elif st[i] >= 'a' and st[i] <= 'z':
    count_lower += 1
  else:
    count_symbol += 1

print(f"""
Numbers: {count_num}
Lower : {count_lower}
Upper : {count_upper}
Symbol : {count_symbol}
""")
```

3. Write a program that takes a string input and creates a new string by replacing:

- all vowels with '*'
- all digits with '#'
- all spaces with '_'

Display both original and modified strings.

```python
st = input("Enter String: ")
orignal_st = st
st = list(st)

vowels = ['a', 'e', 'i', 'o', 'u']

for i in range(len(st)):
  if st[i] in vowels:
    st[i] = '*'
  elif st[i] >= '0' and st[i] <= '9':
    st[i] = '#'
  elif st[i] == ' ':
    st[i] = '_'

st = ''.join(st)

print(f"""
Orignal String: {orignal_st}
Modified String: {st}
""")
```
